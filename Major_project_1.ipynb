```python
indices = []
for i,row in d4_sorted.iterrows():
    indices.append(i)
```

```python
len(indices)
```

##sequencial duplicates with less than 2 words

```python
import itertools
stage1_dedupe_asins = []
i = 0
j = 0
num_data_points = d4_sorted.shape[0]
while i < num_data_points and j < num_data_points:
    
    previous_i = i
    a = d4_sorted['title'].loc[indices[i]].split()
    j = i+1
    while j < num_data_points:
        b = d4['title'].loc[indices[j]].split()
        length = max(len(a), len(b))
        count  = 0
        for k in itertools.zip_longest(a,b): 
            if (k[0] == k[1]):
                count += 1
        if (length - count) > 2:
            stage1_dedupe_asins.append(d4_sorted['asin'].loc[indices[i]])
            if j == num_data_points-1: 
              stage1_dedupe_asins.append(d4_sorted['asin'].loc[indices[j]])
            i = j
            break
        else:
            j += 1
    if previous_i == i:
        break
```

```python
d5 = d4.loc[d4['asin'].isin(stage1_dedupe_asins)]
```

```python
d5.shape
```

```python
d5.to_pickle('17k_apperal_data')
```

```python
indices = []
for i,row in d5.iterrows():
    indices.append(i)

stage2_dedupe_asins = []
while len(indices)!=0:
    i = indices.pop()
    stage2_dedupe_asins.append(d5['asin'].loc[i])
    a = d5['title'].loc[i].split()
    for j in indices:
        
        b = d5['title'].loc[j].split()
        length = max(len(a),len(a))
        count  = 0
        for k in itertools.zip_longest(a,b): 
            if (k[0]==k[1]):
                count += 1
        if (length - count) < 3:
            indices.remove(j)
```
